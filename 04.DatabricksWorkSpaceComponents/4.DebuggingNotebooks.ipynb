```python
# Calculte the final price of an item and print the amount
item_price = 120.00
tax_rate = 20
discount = 10.00

# 1. Calculate the tax to be applied
tax = item_price *(tax_rate/100)

# 2. Calculate the discounted price
discounted_price = item_price - discount

# 3. Calculate the final price
final_price = discounted_price + tax

# 4. Print the final price
print(f"The final price is {final_price}")
```
