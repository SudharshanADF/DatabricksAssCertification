#### File System utilities
- %fs: used for adhoc file system operations
- dbutils.fs used for programatical operations

```python
%fs ls /
```

```python
# dbutils.fs.ls('/')
display(dbutils.fs.ls('/'))
```

```python
items = dbutils.fs.ls('/databricks-datasets/')
```

```python
folderCount = len([item for item in items if item.name.endswith('/')])
print(f"Total Folders: {folderCount}")
fileCount = len([item for item in items if not item.name.endswith('/')])
print(f"Total Files: {fileCount}")
```
